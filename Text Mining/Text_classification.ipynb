```python
import pandas as pd

complaint_df = pd.read_csv('/content/complaint1700.csv')
noncomplaint_df = pd.read_csv('/content/noncomplaint1700.csv')
test_df = pd.read_csv('/content/test_data.csv')

print("Complaint Data:")
display(complaint_df.head())

print("\nNon-Complaint Data:")
display(noncomplaint_df.head())

print("\nTest Data:")
display(test_df.head())
```

```python
print("Complaint DataFrame Info:")
complaint_df.info()

print("\nComplaint DataFrame Missing Values:")
print(complaint_df.isnull().sum())

print("\nNon-Complaint DataFrame Info:")
noncomplaint_df.info()

print("\nNon-Complaint DataFrame Missing Values:")
print(noncomplaint_df.isnull().sum())
```

```python
combined_df = pd.concat([complaint_df, noncomplaint_df], ignore_index=True)
display(combined_df.head())
print("\nShape of combined_df:", combined_df.shape)
```

```python
import re

def clean_text(tweet):
    # Remove user mentions
    tweet = re.sub(r'@\w+', '', tweet)
    # Remove URLs
    tweet = re.sub(r'http\S+|https\S+', '', tweet)
    # Remove special characters, punctuation, and numbers, keeping only letters and spaces
    tweet = re.sub(r'[^a-zA-Z\s]', '', tweet)
    # Convert to lowercase
    tweet = tweet.lower()
    return tweet

combined_df['cleaned_tweet'] = combined_df['tweet'].apply(clean_text)
test_df['cleaned_tweet'] = test_df['tweet'].apply(clean_text)

print("Combined DataFrame after cleaning:")
display(combined_df[['tweet', 'cleaned_tweet']].head())

print("\nTest DataFrame after cleaning:")
display(test_df[['tweet', 'cleaned_tweet']].head())
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

# Get English stop words
stop_words = set(stopwords.words('english'))

# Define a function to tokenize and remove stop words
def tokenize_and_remove_stopwords(text):
    if isinstance(text, str):
        tokens = word_tokenize(text)
        filtered_tokens = [word for word in tokens if word not in stop_words]
        return filtered_tokens
    return []

# Apply the function to the cleaned_tweet column
combined_df['tokens'] = combined_df['cleaned_tweet'].apply(tokenize_and_remove_stopwords)
test_df['tokens'] = test_df['cleaned_tweet'].apply(tokenize_and_remove_stopwords)

# Display the head of both dataframes
print("Combined DataFrame after tokenization and stop word removal:")
display(combined_df[['tweet', 'cleaned_tweet', 'tokens']].head())

print("\nTest DataFrame after tokenization and stop word removal:")
display(test_df[['tweet', 'cleaned_tweet', 'tokens']].head())
```

```python
# The first 1700 rows came from complaint_df
combined_df['label'] = 'non-complaint'
combined_df.loc[:1699, 'label'] = 'complaint'

print("Combined DataFrame with 'label' column:")
display(combined_df.head())

print("\nValue counts of the 'label' column:")
display(combined_df['label'].value_counts())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Instantiate TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit on combined_df and transform
X_train_tfidf = tfidf_vectorizer.fit_transform(combined_df['cleaned_tweet'])

# Transform test_df
X_test_tfidf = tfidf_vectorizer.transform(test_df['cleaned_tweet'])

print("Shape of X_train_tfidf:", X_train_tfidf.shape)
print("Shape of X_test_tfidf:", X_test_tfidf.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB

# Instantiate a MultinomialNB object
nb_model = MultinomialNB()

# Train the Naive Bayes model
nb_model.fit(X_train_tfidf, combined_df['label'])

print("Naive Bayes model trained successfully.")
```

## Predict on test data



```python
y_pred = nb_model.predict(X_test_tfidf)
print("First 10 predictions on the test data:")
print(y_pred[:10])
```
